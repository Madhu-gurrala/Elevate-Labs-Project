## Logistic Regression : Predict the Probability of Return.


```python
# Creating a target variable
e_commerce['Returned']=(e_commerce['Return Rate']>10.4).astype(int)

# Features and label
X=e_commerce[['Price','Discount','Tax Rate','Stock Level','Shipping Cost','Popularity Index']]
y=e_commerce['Returned']

# Splitting
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=34,stratify=y)
# Train Model
model=LogisticRegression()
model.fit(X_train,y_train)

# Predict probabilities
e_commerce['Return_Probability']=model.predict_proba(X)[:,1]
print(e_commerce[['Product ID','Returned','Return_Probability']])
```

## Save high risk products to csv

```python
# Products with >60% probability
high_risk=e_commerce[e_commerce['Return_Probability']>0.5053]

# save to csv
high_risk.to_csv('high_risk_products.csv',index=False)
print(high_risk[['Product ID', 'Product Name', 'Category', 'Return_Probability']])
```
